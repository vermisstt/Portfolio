1\. File `"bank_customers.csv"` that you can find at `"https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"` describes bank clinets. It contains three columns: customer age (`CustAge`), period of relationship with the bank in months (`Months`), and credit limit (`CredLim`). Plot a histogram for the column `Month` and compute its mean, median and the mode. Show these values on the histogram. Make a second copy of the histogram, compute quartiles and show them on the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/bank_customers.csv" #our first step is to load data
df = pd.read_csv(url)
months = df['Months']

plt.figure(figsize=(10, 6)) #histogram - mean, median and mode
plt.hist(months, alpha=0.7, color='brown')
plt.title('Histogram 1')
plt.xlabel('Months')
plt.ylabel('Frequency')
mean = months.mean()
median = months.median()
mode = stats.mode(months,keepdims=True).mode[0]
plt.axvline(mean, color='purple', linestyle='--', label=f'Mean = {mean:.2f}')
plt.axvline(median, color='blue', linestyle='--', label=f'Median = {median:.2f}')
plt.axvline(mode, color='yellow', linestyle='--', label=f'Mode = {mode:.2f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
plt.figure(figsize=(10, 6)) #quartiles
plt.hist(months, alpha=0.7, color='yellow')
plt.title('Histogram 2')
plt.xlabel('Months')
plt.ylabel('Frequency')
q1 = months.quantile(0.25)
q2 = months.quantile(0.5)
q3 = months.quantile(0.75)
plt.axvline(q1, color='red', linestyle='--', label=f'Q1 = {q1:.2f}')
plt.axvline(q2, color='orange', linestyle='--', label=f'Q2 (Median) = {q2:.2f}')
plt.axvline(q3, color='purple', linestyle='--', label=f'Q3 = {q3:.2f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

It contains three columns: customer age (`CustAge`), period of relationship with the bank in months (`Months`), and credit limit (`CredLim`). Plot a histogram for the column `CustAge` and compute its standard deviation and the interquartile range. Show them on the plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/bank_customers.csv"
df = pd.read_csv(url)
mean = df['CustAge'].mean()
std = df['CustAge'].std()
iqr = df['CustAge'].quantile(0.75) - df['CustAge'].quantile(0.25)
plt.hist(df['CustAge'], bins=20, color='purple')
plt.axvline(mean, color='r', label='Mean')
plt.axvline(mean + std, color='orange')
plt.axvline(mean - std, color='orange')
plt.axvline(df['CustAge'].quantile(0.25), color='blue')
plt.axvline(df['CustAge'].quantile(0.75), color='green')
plt.text(mean + std, 0.5*plt.ylim()[1], f'Std = {std:.2f}',color='black')
plt.text(mean - std, 0.5*plt.ylim()[1], f'Std = {std:.2f}', color='pink')
plt.text((df['CustAge'].quantile(0.25)+df['CustAge'].quantile(0.75))/2, 0.8*plt.ylim()[1], f'IQR = {iqr:.2f}', color='darkgreen', ha='center')
plt.xlabel('CustAge')
plt.title('CustAge')
plt.legend()
plt.show()
```
