1\. Create a plot with graphs of two functions: $y=e^{-x^2}$ and $y=3x^2 e^{-x^2}$. Add x and y axes labels and the legend. Turn on the grid. Customize curve colors and styles.

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating x-values from -3 to 3 with 200 points
x = np.linspace(-3, 3, 200)
# Calculating y values for both functions
y1 = np.exp(-x**2)               # y = e^(-x²)
y2 = 3 * x**2 * np.exp(-x**2)    # y = 3x²e^(-x²)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, color='pink', linestyle='-', linewidth=2, label=r'$y = e^{-x^2}$')
plt.plot(x, y2, color='purple', linestyle='--', linewidth=2, label=r'$y = 3x^2e^{-x^2}$')
# Adding labels and title
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Function Comparison', fontsize=14)
plt.legend(fontsize=11)
# Customizing grid
plt.grid(True, linestyle=':', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

1\. Create 3D surface plot of a function $z=x^2-y^2$. Add axes labels and color bar.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2

# Creating the figure and 3D axis
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
# Creating the surface plot
surf = ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='black')
# Adding labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of $z = x^2 - y^2$')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5, label='Z value')
ax.view_init(elev=30, azim=30)
plt.tight_layout()
plt.show()
```

4\. Again take the dataset with movies popularity. Find 3 the largest and 3 the smallest popularity values and create a new array where 0 correspond to these three less popular movies, 2 are assigned to the three most popular movies and all other are marked by 1. Plot 3D scatter plot where marker colors are assigned according to this array.

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
file_id = '183MmHwoozSoziFt4o0BJvl-RRf3PAGcC'
url = f'https://drive.google.com/uc?id={file_id}'
df = pd.read_csv(url)
```

```python
# cleaning data
df = df.dropna(subset=['popularity', 'vote_average', 'vote_count'])
df = df[df['vote_count'] > 100]

# defining best and worst movies
top3, bottom3 = df.nlargest(3, 'popularity'), df.nsmallest(3, 'popularity')

#adjusting classification
df['popularity_class'] = 1
df.loc[bottom3.index, 'popularity_class'] = 0
df.loc[top3.index, 'popularity_class'] = 2
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

# establishing variables
x = df['vote_average']
y = np.log10(df['vote_count'])
z = df['popularity']
c = df['popularity_class']
sc = ax.scatter(x, y, z, c=c, cmap='viridis', alpha=0.7, s=30, edgecolors='w')
ax.set_xlabel('Average Rating', labelpad=15)
ax.set_ylabel('Vote Count', labelpad=15)
ax.set_zlabel('Popularity', labelpad=15)
ax.set_title('Movie Popularity', pad=20)

# adjusting color scale
cbar = plt.colorbar(sc, ticks=[0,1,2])
cbar.ax.set_yticklabels(['Least Popular', 'Medium', 'Most Popular'])
cbar.set_label('Popularity Classification', rotation=270, labelpad=15)
ax.grid(True)
ax.view_init(elev=20, azim=45)
plt.tight_layout()
plt.show()
print("\n3 Most Popular Movies:")
print(top3[['title', 'popularity']])
print("\3 Less Popular Movies:")
print(bottom3[['title', 'popularity']])
```

File `"chess_games.csv"`  that you can find at `"https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"` contains a set of just over 20,000 games collected from a selection of users on the site `Lichess.org` (taken from public dataset collection at `https://www.kaggle.com/datasnaek/chess`).

Read this file and take columns "turns", "white_rating" and "black_rating". Use `abs("white_rating"-"black_rating")` (i.e., the absolute value of difference between columns "white_rating" and "black_rating") as x variable and column "turns" as y variable. Create 2D histogram for x and y variables that demonstrates how a number of game turns depends on the difference between the player rates. Add reasonable axes labels.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/chess_games.csv"
df = pd.read_csv(url, comment='#')
print(df.head())
```

```python
# checking the columns
required_cols = ['turns', 'white_rating', 'black_rating']

df = df.dropna(subset=required_cols)
df['rating_diff'] = (df['white_rating'] - df['black_rating']).abs()

plt.figure(figsize=(10, 6))
plt.hist2d(df['rating_diff'], df['turns'], bins=50, cmap='autumn_r', cmin=1)
plt.colorbar(label='Number of games')
plt.xlabel('Rating difference between players')
plt.ylabel('Number of moves')
plt.title('Chess Game Length vs Rating Difference')
plt.xlim(0, 1000)
plt.ylim(0, 200)
plt.show()
```
